## Convert Raw Restaurant Data to a Clean Dataframe

Code to generate a dataframe with nutritional facts for every item at each fast food chain.

```python
import pandas as pd
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

import matplotlib as plt
%matplotlib inline

import re
from pandas.io.json import json_normalize

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

### Import raw data

```python
df_mcd_raw = pd.read_json(r'C:\Users\Jenny\Desktop\Professional\Python\Practice\Fast_Food\mcd.json')
df_wendys_raw = pd.read_json(r'C:\Users\Jenny\Desktop\Professional\Python\Practice\Fast_Food\wendys.json')
df_bk_raw = pd.read_json(r'C:\Users\Jenny\Desktop\Professional\Python\Practice\Fast_Food\bk.json')
df_tb_raw = pd.read_json(r'C:\Users\Jenny\Desktop\Professional\Python\Practice\Fast_Food\tb.json')
df_wendys_raw.head(1)
```

### Create a dataframe and clean the data

```python
# Create a dataframe for each restaurant
df_mcd = pd.concat([pd.DataFrame(json_normalize(x)) for x in df_mcd_raw['hits']],ignore_index=True)
df_wendys = pd.concat([pd.DataFrame(json_normalize(x)) for x in df_wendys_raw['hits']],ignore_index=True)
df_bk = pd.concat([pd.DataFrame(json_normalize(x)) for x in df_bk_raw['hits']],ignore_index=True)
df_tb = pd.concat([pd.DataFrame(json_normalize(x)) for x in df_tb_raw['hits']],ignore_index=True)
df_mcd.head()
```

```python
# Merge dataframes to create one dataframe with all restaurants
df_menu = df_mcd.append([df_wendys, df_bk, df_tb])
df_menu.head()
```

```python
# Remove unused columns, rename columns, format the 'last_updated' column
df_menu = df_menu.drop(['_id', '_index', '_score', '_type'], axis=1)
df_menu = df_menu.rename(columns = lambda x: re.sub('fields\.|nf_','',x))
df_menu = df_menu.rename(columns = {'updated_at':'last_updated','total_carbohydrate':'carbs','sugars':'sugar'})
df_menu['last_updated'] = df_menu['last_updated'].apply(lambda x: str(x)[0:10])
df_menu.head(1)
```

```python
# Remove items that are not from the correct restaurant and remove duplicate items
is_restaurant = ["McDonald's", "Wendy's", "Burger King", "Taco Bell"]
df_menu = df_menu[df_menu['brand_name'].isin(is_restaurant)]
df_menu = df_menu.drop_duplicates(['item_id'])
df_menu = df_menu.drop(df_menu[df_menu['item_id'] == '59213b3ffb693b8f689482a0'])
df_menu = df_menu.drop(df_menu[df_menu['item_id'] == 'c6408349c3c41ea357fd1585'])
df_menu
```

### Categorize items as Entree, Side, Dessert, Drink

```python
side_hits = "mccheddar|mcbites|fries|Fries|yogurt|apple slices|apples slices|cuties|side|hash|hashbrown|pretzel|tots|mac n|soup|dippers|rice|chips and|snack|potatoes|baked potato|chili,|rings|sticks|nachos|beans|salsa|mini|roll up|tortilla chips|pintos|curds|rice"
entree_hits = "triple|panini|Croissan'Wich|blt|chicken jr|chicken club|chicken breast|big breakfast|double|mcdouble|mcchicken|deluxe|tenders|nuggets|artisan|buttermilk|mcnuggets|selects|naked|sourdough|burger|cheeseburger|patty|cod|dog|rodeo|griller|quesadilla|sausage|stack|sandwich|taco|chalupa|wrap|box|meximelt|quesarito|tostada|gordita|crunchwrap|bowl|biscuit|burrito|platter|salad|mcmuffin|oatmeal|bagel|menuouble|mcgriddle|pounder|lobster|filet|pizza|big mac|mcrib|baconator"
dessert_hits = "shake|mcflurry|sundae|cookie|pie|cone|ice cream|cake|cheesecake|muffin|croissant|sherbet|pastry|melts|frosty|roll|cinnabon|caramel apple|twists"
drink_hits = ", extra small|, small|, medium|, large|, kids|, value|punch|cocoa|box|juice|slushie|soft serve|smoothie|water|freeze|jug|tea|cola|mocha|macchiato|oz|ml|dew|proof|coffee|milk"

# entrees: juice box, side salad, oatmeal raisin cookie

df_menu['item_type'] = df_menu['item_name'].apply(lambda x: 
    'entree' if re.search(entree_hits, x, re.IGNORECASE)
    else 'side' if re.search(side_hits, x, re.IGNORECASE) 
    else 'dessert' if re.search(dessert_hits, x, re.IGNORECASE)
    else 'drink' if re.search(drink_hits, x, re.IGNORECASE)
    else 'misc'
    )

# Add 'tag' column for tagging item in next step
df_menu['tag'] = 'none'

# Correctly categorize outlier items
# juice box
df_menu.loc[df_menu['item_id'] == '513fc9e73fe3ffd40300113a', 'item_type'] = 'drink'
# side salad
df_menu.loc[df_menu['item_id'] == '513fc9e73fe3ffd40300117f', 'item_type'] = 'side'
# oatmeal raisin cookie
df_menu.loc[df_menu['item_id'] == '513fc9e73fe3ffd40300115c', 'item_type'] = 'dessert'
# snack size McFlurry's
df_menu.loc[df_menu['item_id'] == '513fc9e73fe3ffd403001184', 'item_type'] = 'dessert'
df_menu.loc[df_menu['item_id'] == '513fc9e73fe3ffd403001183', 'item_type'] = 'dessert'
# muffin
df_menu.loc[df_menu['item_id'] == 'c6408349c2be5957d572427f', 'item_type'] = 'dessert'
# syrup
df_menu.loc[df_menu['item_id'] == '513fc9e73fe3ffd4030010fc', 'item_type'] = 'misc'

# Re-order columns            
df_menu = df_menu[["brand_name", "brand_id", "item_name", "item_id", "item_type","tag", "calories", "total_fat", "saturated_fat", "cholesterol", "carbs", "sugar", "protein", "sodium", "dietary_fiber", "servings_per_container", "serving_size_qty", "serving_size_unit", "serving_weight_grams","last_updated"]]
```

### Tag items with keywords

```python
breakfast_tag = "biscuit|breakfast|platter|mcmuffin|oatmeal|bagel|mcgriddle|sausage|bacon"
coffee_tag = "coffee|frappe|latte|mocha|cappuccino|macchiato"
juice_tag = "juice|lemonade|limemade|punch"
other_bev_tag = "tea|hot chocolate|smoothie|powerade|vitamin water"
soda_tag = ", extra small|, small|, medium|, large|cola"

df_menu['tag'] = df_menu['item_name'].apply(lambda x: 
    'breakfast' if re.search(breakfast_tag, x, re.IGNORECASE) 
    else 'coffee' if re.search(coffee_tag, x, re.IGNORECASE)
    else 'none'
    )
```

```python
# Fix incorrect tags
incorrect_tags = df_menu[df_menu['item_name'].str.contains('Browns, |Rings, |Fries, ')]['item_id']

for x in incorrect_tags:
    df_menu.loc[df_menu['item_id'] == x, 'tag'] = 'none'
    
# Correctly categorize outlier items
# chicken platter
df_menu.loc[df_menu['item_id'] == 'c6408349b8949aca2ca326e0', 'tag'] = 'none'
# coffee cake
df_menu.loc[df_menu['item_id'] == 'c640834987a34da65a58b785', 'tag'] = 'breakfast'
df_menu.loc[df_menu['item_id'] == 'c64083498ed2241e56f001a1', 'tag'] = 'breakfast'
df_menu.loc[df_menu['item_id'] == 'c6408349f98c9dbf9bb5d806', 'tag'] = 'breakfast'
# muffin
df_menu.loc[df_menu['item_id'] == 'c6408349c2be5957d572427f', 'tag'] = 'breakfast'
# breakfast burritos
df_menu.loc[df_menu['item_id'] == 'c6408349b496a192d4b8cb47', 'tag'] = 'breakfast'
df_menu.loc[df_menu['item_id'] == 'c64083492cd1ad2cd10e3bd3', 'tag'] = 'breakfast'
df_menu.loc[df_menu['item_id'] == '32f12700d10315537f6b13ad', 'tag'] = 'breakfast'
```
